```python
import albumentations as A
import augmax
import imageio
import matplotlib.pyplot as plt

import random
```

```python
input_image = imageio.imread('https://github.com/khdlr/augmax/raw/master/docs/teddy.png')
```

```python
ops = [
    A.HorizontalFlip(p=0.5),
    A.Rotate(15),
    A.RandomCrop(width=128, height=128),
    A.RandomBrightnessContrast(),
    A.ToGray()
]

for step in range(len(ops)):
  transform = A.Compose(ops[:step])

  # random.seed(12387)
  random.seed(1728)
  out = transform(image=input_image)['image']
  imageio.imwrite(f'imgs/augmax_sample_step_{step}.png', out)
  plt.imshow(out)
  plt.show()
```
